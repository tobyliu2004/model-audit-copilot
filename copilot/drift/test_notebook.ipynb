```python
import pandas as pd
import numpy as np
from copilot.drift.drift_detector import compare_datasets

# Simulated reference data
np.random.seed(42)
reference_df = pd.DataFrame({
    "age": np.random.normal(40, 10, 1000),
    "income": np.random.normal(50000, 8000, 1000),
})

# Simulated current data with drift
current_df = pd.DataFrame({
    "age": np.random.normal(45, 15, 1000),         # Shifted mean and std
    "income": np.random.normal(50000, 8000, 1000)  # No drift here
})

# Run drift detection
report = compare_datasets(reference_df, current_df)

# Print summary and plot
print(report.summary())
report.plot()
```
